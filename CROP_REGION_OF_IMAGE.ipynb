```
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```
img1 = cv2.imread('SET14_BMP/zebra.bmp')
img=cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)

zebra = img[200:300, 107:194] #THIS IS THE ROI = REGION OF IMAGE
img[0:100, 0:87] = zebra  #100 = 300 - 100 AND 87 = 194 - 107 NEW REGION OF IMAGE
cv2.rectangle(img,(116,270),(160,220),(200,0,0),2) # X-AXIS in B/W 116 and 160, Y-Axis in B/W 270 and 220 BELOW RECTANGLE
cv2.rectangle(img,(1,1),(90,98),(0,255,255),2) # X-AXIS in B/W 1 and 90, Y-Axis in B/W 1 and 98 ABOVE RECTANGLE
fig = plt.figure(figsize = (14,14), dpi = 100)
ax = plt.subplot("121")
ax.imshow(img)
plt.grid(0)
plt.show()

ax = plt.subplot("122")
ax.imshow(zebra)
plt.grid(0)
plt.show()
```
